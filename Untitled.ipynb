```python
#Name : Rahma Mohamed Abdo Mohamed Elsharkawy
#Email: rahma.elsharkawy82005@gmail.com

import dimod
Q = {
    (0, 0): 1,
    (1, 1): 2,
    (2, 2): 3,
    (0, 1): 3,
    (0, 2): -1
}
sampler = dimod.SimulatedAnnealingSampler()
response = sampler.sample_qubo(Q, num_reads=100)
best_sample = next(response.samples(n=1))
best_energy = response.first.energy
```

```python
print("Best solution:", best_sample)
print("Minimum energy:", best_energy)
```
