```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator, StatevectorSimulator
from qiskit.visualization import plot_histogram
```

```python
import numpy as np
```

```python
#Define the q circuit
q= QuantumRegister (1,'q')
c = ClassicalRegister(1, 'c')
qc = QuantumCircuit(q,c)

#perform operators
angle_of_rotation = np.pi/3

qc.ry(2* angle_of_rotation, q[0])

qc.measure(q,c)

#visual mpl for coloring
qc.draw(output='mpl')
```

```python
#!pip install pylatexenc
```

```python
sim =AerSimulator()

job = sim.run(qc, shots=1000)
results = job.result()
count= job.result().get_counts(qc)
count
```

```python
plot_histogram(count)
```

```python
sv_sim =StatevectorSimulator()
```

```python
job= sv_sim.run(qc)
current_state = job.result().get_statevector(qc)
current_state.data
#j in the output equivelanr to i (imaginary number) 0+0i and 1+0i
```

```python
x_value= current_state[0].real

y_value= current_state[1].real
x_value,y_value
```

```python
#Rewan Mostafa SayedAhmed Ramadan 
#rewanmostafa321@gmail.com 

# Import reqired objects
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator, StatevectorSimulator
from qiskit.visualization import plot_histogram
# define the quantum circuit

q = QuantumRegister (2,'q')
c = ClassicalRegister(2, 'c')
qc = QuantumCircuit(q,c)

# apply the Hadamard operator on the first qubit (q[0])
qc.h(q[0])

# apply the CNOT operator
qc.cx(q[0] , q[1])

# measure the quantum bit, and store the outcome on the classical bit
qc.measure(q,c)
qc.draw("mpl")
```

```python
# viualize the quntum circuit 
sim =AerSimulator()
job = sim.run(qc, shots=1000)
results = job.result()
counts= job.result().get_counts(qc)
plot_histogram(counts)
```
